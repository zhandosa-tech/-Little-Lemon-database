```python
import mysql.connector
from mysql.connector import pooling
from mysql.connector import Error

dbconfig = {
    "host": "127.0.0.1",
    "user": "root",
    "password": "Geometry123",
    "database": "little_lemon_db"
}

try:
    # Establish a connection pool with 2 connections
    pool = pooling.MySQLConnectionPool(
        pool_name="pool_b",
        pool_size=2,
        **dbconfig
    )
    print(self_reply := f"Connection pool '{pool.pool_name}' created successfully with size {pool.pool_size}.")

except Error as e:
    print(f"Error while creating connection pool: {e}")
```

```python
# Guest reservation data inputs
guests = [
    {"table": 8, "first": "Anees", "last": "Java", "time": "18:00:00", "emp_id": 6},
    {"table": 5, "first": "Bald", "last": "Vin", "time": "19:00:00", "emp_id": 6},
    {"table": 12, "first": "Jay", "last": "Kon", "time": "19:30:00", "emp_id": 6}
]

connections = []

try:
    # 1. Acquire connections for Guest 1 and Guest 2
    print("Acquiring connections for Guest 1 and Guest 2...")
    connections.append(pool.get_connection())
    connections.append(pool.get_connection())
    
    # 2. Add a connection dynamically for Guest 3 to prevent starvation
    print("Pool capacity reached. Dynamically adding a connection for Guest 3...")
    pool.add_connection()
    connections.append(pool.get_connection())
    
    # 3. Insert reservation details into the Bookings table
    insert_query = """
    INSERT INTO Bookings (TableNumber, FirstName, LastName, BookingSlot, EmployeeID)
    VALUES (%s, %s, %s, %s, %s);
    """
    
    for i, guest in enumerate(guests):
        cursor = connections[i].cursor()
        cursor.execute(insert_query, (guest["table"], guest["first"], guest["last"], guest["time"], guest["emp_id"]))
        connections[i].commit()
        cursor.close()
        print(f"Successfully inserted booking for {guest['first']} {guest['last']}.")

except Error as e:
    print(f"Database operation failed: {e}")

# 4. Attempt to return all 3 connections back to a pool configured for 2
print("\nReturning connections back to the pool...")
for idx, conn in enumerate(connections):
    try:
        conn.close()
        print(f"Connection {idx + 1} successfully returned to the pool.")
    except pooling.PoolError as pe:
        print(f"PoolError on Connection {idx + 1}: {pe}")
    except Exception as ex:
        print(f"An unexpected error occurred while closing connection {idx + 1}: {ex}")
```

```python
try:
    connection = pool.get_connection()
    cursor = connection.cursor()

    # Query 1: Manager details
    cursor.execute("SELECT Name, EmployeeID FROM Employees WHERE Role = 'Manager';")
    manager = cursor.fetchone()
    print(f"--- Little Lemon Manager ---\nName: {manager[0]}, ID: {manager[1]}\n")

    # Query 2: Highest salary earner
    cursor.execute("SELECT Name, Role FROM Employees ORDER BY Salary DESC LIMIT 1;")
    highest_earner = cursor.fetchone()
    print(f"--- Top Earner ---\nName: {highest_earner[0]}, Role: {highest_earner[1]}\n")

    # Query 3: Bookings between 18:00 and 20:00
    cursor.execute("SELECT COUNT(*) FROM Bookings WHERE BookingSlot BETWEEN '18:00:00' AND '20:00:00';")
    booking_count = cursor.fetchone()
    print(f"--- Peak Hours ---\nBookings between 18:00 and 20:00: {booking_count[0]}\n")

    # Query 4: Waiting guests sorted by booking slot
    query_waiting = """
    SELECT BookingID, CONCAT(FirstName, ' ', LastName) AS FullName 
    FROM Bookings 
    ORDER BY BookingSlot ASC;
    """
    cursor.execute(query_waiting)
    print("--- Guest Seating Queue ---")
    for row in cursor.fetchall():
        print(f"Booking ID: {row[0]} | Guest Name: {row[1]}")

except Error as e:
    print(f"Report generation failed: {e}")
finally:
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
```

```python
try:
    connection = pool.get_connection()
    cursor = connection.cursor()

    # Drop procedure if it already exists to avoid conflicts
    cursor.execute("DROP PROCEDURE IF EXISTS BasicSalesReport;")

    # Construct the Stored Procedure text code block
    stored_procedure_query = """
    CREATE PROCEDURE BasicSalesReport()
    BEGIN
        SELECT 
            SUM(BillAmount) AS Total_Sales,
            AVG(BillAmount) AS Average_Sale,
            MIN(BillAmount) AS Min_Bill_Paid,
            MAX(BillAmount) AS Max_Bill_Paid
        FROM Orders;
    END
    """
    cursor.execute(stored_procedure_query)
    print("Stored procedure 'BasicSalesReport' created successfully.")

    # Execute and test the procedure
    cursor.callproc("BasicSalesReport")
    
    print("\n--- Basic Sales Report Statistics ---")
    for result in cursor.stored_results():
        report = result.fetchone()
        print(f"Total Sales:     ${report[0]:.2f}")
        print(f"Average Sale:    ${report[1]:.2f}")
        print(f"Minimum Bill:    ${report[2]:.2f}")
        print(f"Maximum Bill:    ${report[3]:.2f}")

except Error as e:
    print(f"Stored procedure execution failed: {e}")
finally:
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
```

```python
try:
    connection = pool.get_connection()
    
    # Instantiate a buffered cursor
    cursor = connection.cursor(buffered=True)

    # Join Bookings and Employees to fetch upcoming display details
    kitchen_display_query = """
    SELECT 
        b.BookingSlot, 
        CONCAT(b.FirstName, ' ', b.LastName) AS Guest_Name, 
        e.Name AS Employee_Name, 
        e.Role AS Employee_Role
    FROM Bookings b
    INNER JOIN Employees e ON b.EmployeeID = e.EmployeeID
    ORDER BY b.BookingSlot ASC
    LIMIT 3;
    """
    
    cursor.execute(kitchen_display_query)
    results = cursor.fetchall()

    print("--- Kitchen Display Notification Screen ---")
    for row in results:
        print(f"Slot: [{row[0]}]")
        print(f"Guest: [{row[1]}]")
        print(f"Assigned to: {row[2]} [{row[3]}]\n")

except Error as e:
    print(f"Kitchen display queue retrieval failed: {e}")
finally:
    if 'connection' in locals() and connection.is_connected():
        cursor.close()
        connection.close()
        print("Kitchen session connection successfully returned to pool.")
```
